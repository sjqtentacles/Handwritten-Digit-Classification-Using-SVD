#Using Scikit-Learn's Handwritten Digit Dataset - Classification with SVD

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import datasets

data = datasets.load_digits()
```

The dataset is loaded as a dictionary object. From there, the important items are data['images'] for the matrix for each image, data['data'] for the flattened matrix for each image, and data['target'] for the label of each image (0-9).

```python
import pandas as pd

df = pd.DataFrame(data['data'])
df['label'] = data['target']
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[df.columns[0:-1]],df['label'])

clf = LogisticRegression()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
u,e,v = np.linalg.svd(X_train)
print u.shape
print np.diag(e).shape
print v.shape
```

```python
one = df[df['label']==1][0:175].T #transpose because I want each image to be a column, not a row. 
one_labels = one.iloc[-1]
one = one[0:-1]#remove label of 1
zero = df[df['label']==0][0:175].T
zero_labels = zero.iloc[-1]
zero = zero[0:-1]#remove label of 0
two = df[df['label']==2][0:175].T
two_labels = two.iloc[-1]
two = two[0:-1]
three = df[df['label']==3][0:175].T
three_labels = three.iloc[-1]
three = three[0:-1]

print len(one.columns)
print len(one)
print len(zero.columns)
print len(zero)
print len(two.columns)
print len(two)

cutoff = 150

one_train = one[one.columns[0:cutoff]]
one_test = one[one.columns[cutoff:]]
zero_train = zero[zero.columns[0:cutoff]]
zero_test = zero[zero.columns[cutoff:]]
two_train = two[two.columns[0:cutoff]]
two_test = two[two.columns[cutoff:]]
three_train = three[three.columns[0:cutoff]]
three_test = three[three.columns[cutoff:]]

print len(one_train.columns)
```

```python
zu, ze, zv = np.linalg.svd(zero_train)
ou, oe, ov = np.linalg.svd(one_train)
tu, te, tv = np.linalg.svd(two_train)
tru, tre, trv = np.linalg.svd(three_train)
```

```python
def classifyDigit(imageVector):
    zeroVal = np.linalg.norm((np.identity(len(zu))[:k,:] - (zu[:k,:]*(zu[:,:k].T))).dot(imageVector),ord=2)/1.0/np.linalg.norm(imageVector)
    oneVal = np.linalg.norm((np.identity(len(ou))[:k,:] - (ou[:k,:]*(ou[:,:k].T))).dot(imageVector),ord=2)/1.0/np.linalg.norm(imageVector)
    twoVal = np.linalg.norm((np.identity(len(tu))[:k,:] - (tu[:k,:]*(tu[:,:k].T))).dot(imageVector),ord=2)/1.0/np.linalg.norm(imageVector)
    threeVal = np.linalg.norm((np.identity(len(tru))[:k,:] - (tru[:k,:]*(tru[:,:k].T))).dot(imageVector),ord=2)/1.0/np.linalg.norm(imageVector)

    vals = [zeroVal,oneVal,twoVal,threeVal]
    return vals.index(min(vals))

for i in range(1,65,8):
    print 'Using K=',i
    k = i
    zeros_pred = []
    ones_pred = []
    twos_pred = []
    threes_pred = []
    for i in range(len(one_test.columns)):
        zeros_pred.append(classifyDigit(zero_test.T.iloc[i]))
        ones_pred.append(classifyDigit(one_test.T.iloc[i]))
        twos_pred.append(classifyDigit(two_test.T.iloc[i]))
        threes_pred.append(classifyDigit(three_test.T.iloc[i]))

    zero_correct = zeros_pred.count(0)/1.0/len(zeros_pred)
    one_correct = ones_pred.count(1)/1.0/len(ones_pred)
    two_correct = twos_pred.count(2)/1.0/len(twos_pred)
    three_correct = threes_pred.count(3)/1.0/len(threes_pred)
    print "Correct Prediction Percentages"
    print "0s - ",zero_correct
    print "1s - ",one_correct
    print "2s - ",two_correct
    print "3s - ",three_correct
    print "Accuracy - ",(zero_correct+one_correct+two_correct+three_correct)/4.0
    print ''
```
